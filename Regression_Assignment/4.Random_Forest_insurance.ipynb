```python
import pandas as pd
```

```python
Dataset=pd.read_csv("insurance_pre.csv")
Dataset
```

```python
Dataset.columns
```

```python
# Here categorical data(state) is available in the dataset which is converted to numerical data 
# As the categorical data available is nominal, one hot coding method is used to convert it into numerical data
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
Dataset.columns
```

```python
#Input and output split
independent=Dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
independent
```

```python
Dependent=Dataset[["charges"]]
Dependent
```

```python
# Split train and test data using train_test_split function from sklearn 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,Dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
#model creation
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(criterion='squared_error',max_depth=4,max_features=None,n_estimators=100,random_state=0)
regressor.fit(X_train,Y_train)
```

```python
y_predict=regressor.predict(X_test)
```

```python
# evaluating the model performance
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predict)
```

```python
r_score
```
