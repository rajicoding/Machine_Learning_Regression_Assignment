```python
import pandas as pd
Dataset=pd.read_csv("insurance_pre.csv")
```

```python
Dataset
```

```python
Dataset.columns
```

```python
# Here categorical data(state) is available in the dataset which is converted to numerical data 
# As the categorical data available is nominal, one hot coding method is used to convert it into numerical data
Dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True)
Dataset
```

```python
# input and output split 
independent=Dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
independent
```

```python
Dependent=Dataset[["charges"]]
Dependent
```

```python
# Split train and test data using train_test_split function from sklearn 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,Dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
Y_train
```

```python
# model creation using LinearRegression from linear_model of sklearn
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# weight calculated is determined using coef_ function of regressor
weight=regressor.coef_
weight
```

```python
# bias calcuated is determined using intercept_ function of regressor
bias=regressor.intercept_
bias
```

```python
y_predict=regressor.predict(X_test)
```

```python
# evaluation metrics to determine the model performance
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_predict)
```
